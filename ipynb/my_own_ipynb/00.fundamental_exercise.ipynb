```python
tf.random.set_seed(66)
tensor_five=tf.constant(tf.random.uniform(shape=[10],dtype=tf.dtypes.float16,seed=66))
tensor_five
```

```python
print ("The Aggression of tensor_five : \n")
print ("The minimum of tensor_five :"             , tf.reduce_min(tensor_five).numpy())
print ("The maximum of tensor_five :"             , tf.reduce_max(tensor_five).numpy())
print ("The mean of tensor_five :"                , tf.reduce_mean(tensor_five).numpy())
print ("The sum of tensor_five :"                 , tf.reduce_sum(tensor_five).numpy())
print ("The variance of tensor_five :"            , tf.math.reduce_variance(tensor_five).numpy())
print ("The standard deviation of tensor_five :"  , tf.math.reduce_std(tensor_five).numpy())
print ("The position of min value in tensor_five : \n" , tf.argmin(tensor_five))
print ("The position of max value in tensor_five : \n" , tf.argmax(tensor_five))
print ("The position of min value in tensor_five with np.argmin:" , np.argmin(tensor_five))
print ("The position of max value in tensor_five with np.argmax:" , np.argmax(tensor_five))
```
